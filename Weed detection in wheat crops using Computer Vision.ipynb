```python
import zipfile
import os

# Path to your zip file
zip_file_path = 'Multi-modal and Multi-view Image Dataset for Weeds Detection in Wheat Field (MMDWWF).zip'
extract_folder = 'extracted_dataset'

# Extracting the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)

print(f"Dataset extracted to {extract_folder}")
```

```python
# List files in the extracted folder
extracted_files = os.listdir(extract_folder)
print(extracted_files)
```

```python
import zipfile
import os

# Path to the zip files
modal_zip_path = './extracted_dataset/7tg28xns3m-1/multi-modal dataset.zip'
view_zip_path = './extracted_dataset/7tg28xns3m-1/multi-view dataset.zip'

# Extraction folders
modal_extract_folder = './extracted_dataset/modal_data'
view_extract_folder = './extracted_dataset/view_data'

# Extracting the multi-modal dataset ZIP file
with zipfile.ZipFile(modal_zip_path, 'r') as zip_ref:
    zip_ref.extractall(modal_extract_folder)
    print(f"Multi-modal dataset extracted to {modal_extract_folder}")

# Extracting the multi-view dataset ZIP file
with zipfile.ZipFile(view_zip_path, 'r') as zip_ref:
    zip_ref.extractall(view_extract_folder)
    print(f"Multi-view dataset extracted to {view_extract_folder}")
```
