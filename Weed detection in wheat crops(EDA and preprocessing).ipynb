```python
from PIL import Image
import os

# Function to resize images to target size
def resize_images(input_folder, output_folder, target_size):
    os.makedirs(output_folder, exist_ok=True)
    for filename in os.listdir(input_folder):
        image_path = os.path.join(input_folder, filename)
        img = Image.open(image_path)
        img_resized = img.resize(target_size)
        img_resized.save(os.path.join(output_folder, filename))

# Resize weed and wheat images to 512x512
resize_images('./extracted_dataset/wdataset/weed', './resized_dataset/weed', (512, 512))
resize_images('./extracted_dataset/wdataset/wheat', './resized_dataset/wheat', (512, 512))
```

```python
import cv2
import matplotlib.pyplot as plt

# Plot pixel intensity distribution
def plot_pixel_intensity(image_path, title):
    img = cv2.imread(image_path)
    colors = ('b', 'g', 'r')
    for i, col in enumerate(colors):
        hist = cv2.calcHist([img], [i], None, [256], [0, 256])
        plt.plot(hist, color=col)
        plt.xlim([0, 256])
    plt.title(title)
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.show()

# Example: Analyze one image from each class
plot_pixel_intensity('./resized_dataset/weed/1.jpg', 'Weed Image Pixel Intensity')
plot_pixel_intensity('./resized_dataset/wheat/wheat1.png', 'Wheat Image Pixel Intensity')
```

```python
import cv2
import numpy as np
import os

# Function to normalize images
def normalize_image(image_path, output_path):
    image = cv2.imread(image_path)
    normalized_image = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX)
    cv2.imwrite(output_path, normalized_image)

# Normalize all images in the weed and wheat folders
input_folder_weed = './resized_dataset/weed'
output_folder_weed = './normalized_dataset/weed'
input_folder_wheat = './resized_dataset/wheat'
output_folder_wheat = './normalized_dataset/wheat'

os.makedirs(output_folder_weed, exist_ok=True)
os.makedirs(output_folder_wheat, exist_ok=True)

for folder, output_folder in [(input_folder_weed, output_folder_weed), (input_folder_wheat, output_folder_wheat)]:
    for filename in os.listdir(folder):
        if filename.endswith(('.jpg', '.png')):
            normalize_image(os.path.join(folder, filename), os.path.join(output_folder, filename))
```

```python
import matplotlib.pyplot as plt

# Function to calculate average color channel values
def analyze_color_channels(folder_path, label):
    red_vals, green_vals, blue_vals = [], [], []
    for filename in os.listdir(folder_path):
        if filename.endswith(('.jpg', '.png')):
            image = cv2.imread(os.path.join(folder_path, filename))
            red_vals.append(np.mean(image[:, :, 2]))
            green_vals.append(np.mean(image[:, :, 1]))
            blue_vals.append(np.mean(image[:, :, 0]))
    
    plt.figure()
    plt.bar(['Red', 'Green', 'Blue'], [np.mean(red_vals), np.mean(green_vals), np.mean(blue_vals)])
    plt.title(f'Average Color Intensity for {label}')
    plt.ylabel('Average Intensity')
    plt.show()

analyze_color_channels(output_folder_weed, 'Weed Images')
analyze_color_channels(output_folder_wheat, 'Wheat Images')
```

```python
import cv2
import os

# Function to adjust the blue channel
def adjust_blue_channel(image):
    b, g, r = cv2.split(image)
    b = cv2.normalize(b, None, 0, 255, cv2.NORM_MINMAX)  # Normalize the blue channel
    adjusted_image = cv2.merge([b, g, r])
    return adjusted_image

# Input and output folder paths
input_folder = 'normalized_dataset/wheat'  # Replace with your actual path
output_folder = 'adjusted_images/wheat'

# Ensure the output directory exists
os.makedirs(output_folder, exist_ok=True)

# Iterate through all images in the input folder
for image_name in os.listdir(input_folder):
    image_path = os.path.join(input_folder, image_name)
    output_path = os.path.join(output_folder, image_name)
    
    # Read and process each image
    image = cv2.imread(image_path)
    if image is not None:
        adjusted_image = adjust_blue_channel(image)
        cv2.imwrite(output_path, adjusted_image)  # Save the adjusted image
        print(f"Saved adjusted image: {output_path}")
    else:
        print(f"Failed to read image: {image_path}")

print(f"All adjusted images saved to {output_folder}.")
```

```python
import os
import shutil

# Paths to the original folders (used for identifying categories)
resized_weed_folder = 'normalized_dataset/weed'
adjusted_wheat_folder = 'adjusted_images/wheat'

# Path to combined folder
combined_folder = 'combined_images'

# Create categorized subfolders inside combined_folder
categorized_weed_folder = os.path.join(combined_folder, 'weed')
categorized_wheat_folder = os.path.join(combined_folder, 'wheat')

os.makedirs(categorized_weed_folder, exist_ok=True)
os.makedirs(categorized_wheat_folder, exist_ok=True)

# Categorize images based on their origin
for image_name in os.listdir(combined_folder):
    if image_name in os.listdir(resized_weed_folder):
        shutil.move(
            os.path.join(combined_folder, image_name),
            os.path.join(categorized_weed_folder, image_name)
        )
    elif image_name in os.listdir(adjusted_wheat_folder):
        shutil.move(
            os.path.join(combined_folder, image_name),
            os.path.join(categorized_wheat_folder, image_name)
        )

print(f"Images categorized into '{categorized_weed_folder}' and '{categorized_wheat_folder}'.")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np

import os
import cv2

# Paths to categorized folders
categorized_weed_folder = 'combined_images/weed'
categorized_wheat_folder = 'combined_images/wheat'

# Output folders for augmented images
augmented_weed_folder = 'augmented_images/weed'
augmented_wheat_folder = 'augmented_images/wheat'

# Create output folders if they don't exist
os.makedirs(augmented_weed_folder, exist_ok=True)
os.makedirs(augmented_wheat_folder, exist_ok=True)

# Define augmentation parameters
datagen = ImageDataGenerator(
    rotation_range=30,           # Rotate images randomly by up to 30 degrees
    width_shift_range=0.2,       # Horizontal translation
    height_shift_range=0.2,      # Vertical translation
    shear_range=0.2,             # Shear transformation
    zoom_range=0.2,              # Random zoom
    horizontal_flip=True,        # Flip images horizontally
    fill_mode='nearest'          # Fill missing pixels after transformation
)

# Function to apply augmentation
def augment_images(input_folder, output_folder, prefix, num_augments=5):
    # Loop through each image in the folder
    for image_name in os.listdir(input_folder):
        if image_name.endswith(('.jpg', '.png')):
            img_path = os.path.join(input_folder, image_name)
            img = cv2.imread(img_path)  # Read the image
            img = img[np.newaxis, ...]  # Add batch dimension for augmentation
            
            # Apply augmentations and save the augmented images
            i = 0
            for batch in datagen.flow(img, batch_size=1, save_to_dir=output_folder, save_prefix=prefix, save_format='jpg'):
                i += 1
                if i >= num_augments:  # Stop after generating specified number of augmentations
                    break

# Perform augmentation on the weed and wheat images
augment_images(categorized_weed_folder, augmented_weed_folder, prefix='weed', num_augments=5)
augment_images(categorized_wheat_folder, augmented_wheat_folder, prefix='wheat', num_augments=5)

print("Data augmentation completed for weed and wheat images.")
```

```python
import os

# Path to augmented images folder
augmented_images_folder = 'augmented_images'

# Initialize counters
total_images = 0
total_size = 0

# Loop through categories (e.g., 'weed', 'wheat')
for category in os.listdir(augmented_images_folder):
    category_folder = os.path.join(augmented_images_folder, category)
    if os.path.isdir(category_folder):
        images = [f for f in os.listdir(category_folder) if f.endswith(('.jpg', '.png'))]
        total_images += len(images)
        total_size += sum(os.path.getsize(os.path.join(category_folder, img)) for img in images)

# Convert size to MB
total_size_mb = total_size / (1024 * 1024)

print(f"Total Number of Images: {total_images}")
print(f"Total Storage Size: {total_size_mb:.2f} MB")
```
